```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
variable = pd.read_csv(r"C:\Users\91878\Desktop\CWC2023.csv", encoding="latin-1")

# Count the occurrences of each team winning the toss
toss_wins = variable['Toss'].value_counts()

# Get the team that won the toss the most
most_toss_wins = toss_wins.idxmax()
max_toss_wins = toss_wins.max()

# Create a bar plot
plt.figure(figsize=(10, 6))
toss_wins.plot(kind='bar', color='skyblue')
plt.title('Number of Toss Wins by Each Team')
plt.xlabel('Teams')
plt.ylabel('Number of Toss Wins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Team that won the toss the most times:", most_toss_wins)
print("Number of times they won the toss:", max_toss_wins)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
variable = pd.read_csv(r"C:\Users\91878\Desktop\CWC2023.csv", encoding="latin-1")

# Filter the dataset for matches where the winning team is the same as the team that won the toss
matches_won_by_toss_winner = variable[variable['Winner'] == variable['Toss']]

# Count the occurrences of each team winning the toss and match
toss_and_match_wins = matches_won_by_toss_winner['Toss'].value_counts()

# Get the team that won the toss and match the most number of times
most_toss_and_match_wins = toss_and_match_wins.idxmax()
max_toss_and_match_wins = toss_and_match_wins.max()

# Create a bar plot
plt.figure(figsize=(10, 6))
toss_and_match_wins.plot(kind='bar', color='lightgreen')
plt.title('Number of Times a Team Won the Toss and the Match')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("Team that won the toss and match the most times:", most_toss_and_match_wins)
print("Number of times they won the toss and match:", max_toss_and_match_wins)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
variable = pd.read_csv(r"C:\Users\91878\Desktop\CWC2023.csv", encoding="latin-1")

# Group the data by team and sum the runs scored in both innings
team_scores = variable.groupby('Winner')[['Innings1_Run', 'Innings2_Run']].sum()

# Calculate the total runs scored by each team
team_scores['Total_Runs'] = team_scores['Innings1_Run'] + team_scores['Innings2_Run']

# Plotting
plt.figure(figsize=(10, 8))
plt.pie(team_scores['Total_Runs'], labels=team_scores.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Runs Scored by Each Team')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.tight_layout()
plt.show()
```
